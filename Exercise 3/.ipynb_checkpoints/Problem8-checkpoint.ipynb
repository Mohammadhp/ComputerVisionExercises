```python
tf_image = tf.placeholder(dtype = tf.float32, shape=(None, None, 3))
resized_image = tf.image.resize_images(tf_image, (256, 256))
image_vectorized = tf.reshape(resized_image, shape=(256 * 256, 3))
image_normalized = image_vectorized - tf.reduce_mean(image_vectorized, axis = 0)

covariance_matrix = tf.matmul(image_normalized, image_normalized, transpose_a=True)

eigen_values, eigen_vectors = tf.self_adjoint_eig(tf.expand_dims(covariance_matrix, 0))
eigen_vectors = tf.transpose(eigen_vectors[0])
eigen_values = tf.transpose(eigen_values)
eigen_vectors_sorted = [eigen_values[:, 2], eigen_values[:, 1], eigen_values
                       []]

output = tf.matmul(image_normalized, eigen_vectors_sorted)

output_image = tf.abs(tf.reshape(output, shape=[-1, 256, 256, 3]))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    eigen_values, eigen_vectors, image_after_pca = sess.run(
        [eigen_values,eigen_vectors,output_image], feed_dict = {tf_image : image})
    image_after_pca = np.squeeze(image_after_pca)
    print(eigen_decomposition)
    print(eigen_vectors)
    print(eigen_values)
    plt.imshow(image_after_pca)
    plt.show()
    
```
